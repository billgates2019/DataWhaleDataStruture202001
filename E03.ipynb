#### 假设一列货运列车共有n节车厢，每节车厢将停放在不同的车站。假定n个车站的编号分别为1至n，货运列车按照第n站至第1站的次序经过这些车站。车厢的编号与它们的目的地相同。为了便于从列车上卸掉相应的车厢，必须重新排列车厢，使各车厢从前至后按编号1至n的次序排列。当所有的车厢都按照这种次序排列时，在每个车站只需卸掉最后一节车厢即可。

#### 编写算法实现火车车厢的重排，模拟具有n节车厢的火车“入轨”和“出轨”过程。

```python
class Node(object):
    def __init__(self,value):
        self.value = value
        self.succ = None  # succ是紧接着的后续一个节点标识

class Stack(object):
    def __init__(self, node=None):
        self.head=node
        
    def push(self,val):
        node=Node(val)
        
        if self.head == None:
            self.head = node
            return
        
        node.succ = self.head
        self.head = node
        
        
    def pop(self):
        if self.head == None:
            return -100
        
        val = self.head.value
        self.head = self.head.succ
        return val        
    
    
    def include(self,val):
        if self.head == None:
            return False
        
        cur = self.head
        while cur :
            if val==cur.value:
                return True    
            cur = cur.succ
        return False
        
        
    def travel(self):
        if self.head == None:
            print("empty stack")
            return
        
        cur=Stack()
        cur = self.head
        while cur :
            print(cur.value,end = " ")
            cur = cur.succ
        print()
        
    
    

    
def stkprint(Train:Stack(),H1:Stack(),H2:Stack(),N):
    print()
    print("#{0}车厢完成重排时:".format(N))
    
    print("列车车厢排列:".format(N),end="")
    Train.travel()
    

    print("H1车厢排列：".format(N),end="")
    H1.travel()
    

    print("H2车厢排列：".format(N),end="")
    H2.travel()
    

    
    
    
def sorting(destk:Stack(),tempstk1:Stack(),tempstk2:Stack(),N):
    if N==1:
        if tempstk1.include(N):
            val=tempstk1.pop()
            destk.push(val)
        else:
            val=tempstk2.pop()
            destk.push(val)
        return
    
    if N!=1:
        if tempstk1.include(N):
            #以N为界，把tempstk1上段转移到tempstk2，N保留在tempstk1栈顶
            val=0
            while val !=N:
                val=tempstk1.pop()
                if val==N:
                    destk.push(val)
                    break
                tempstk2.push(val)

            
        
        elif tempstk2.include(N):
            #以N为界，把tempstk2上段转移到tempstk1，N保留在tempstk2栈顶
            val=0
            while val !=N:
                val=tempstk2.pop()
                if val==N:
                    destk.push(val)
                    break
                tempstk1.push(val)
                
    stkprint(destk,tempstk1,tempstk2,N)
    
    sorting(destk,tempstk1,tempstk2,N-1)
            
            
            

if __name__ == '__main__':
    
    Train=Stack()
    H1=Stack()
    H2=Stack()
    #H3=Stack() #可以不使用
    
    str1=input("将数字1～9以随意次序输入（以空格隔开）：")  #例如输入：5 8 1 7 4 2 9 6 3
    sarr=str1.split(' ')
    maxSN=len(sarr)
    
    for str2 in sarr:
        val=int(str2)
        Train.push(val)

    print("\n进站前列车车厢排列：",end="")
    Train.travel()
    
    
    #把初始堆栈以最大值maxSN为界，分两段转移到H1和H2，maxSN放入H1栈顶
    val=Train.pop()
    while val != -100:
        H1.push(val)
        if val==maxSN:
            break
        val=Train.pop()
    print()

    val=Train.pop()    
    while val != -100:
        H2.push(val)
        val=Train.pop()
    print()
    

    print("列车车厢全部卸除：",end="")
    Train.travel()
    

    print("H1车厢排列：",end="")
    H1.travel()
    

    print("H2车厢排列：",end="")
    H2.travel()
    
    
    sorting(Train,H1,H2,9)
    

    print("\n重排完成时：")
    print("列车车厢排列",end="")
    Train.travel()
    

    print("H1车厢排列：",end="")
    H1.travel()
    

    print("H2车厢排列：",end="")
    H2.travel()
    
    
```
