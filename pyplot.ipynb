```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.linspace(-2, 6, 50)
y1 = x + 3      # 曲线 y1
y2 = 3 - x      # 曲线 y2
plt.figure()    # 定义一个图像窗口
plt.plot(x, y1) # 绘制曲线 y1
plt.plot(x, y2) # 绘制曲线 y2
plt.show()
```

```python
import math
plt.plot([-5,5],[0,0],'ro')
for i in range(-540,540):
    for j in range(-960,960):
        p=math.sqrt((i+5)*(i+5)+j*j)
        q=math.sqrt((i-5)*(i-5)+j*j)
        if abs(p-q-2)<1:
            plt.plot([i],[j],'ro')
            
```
