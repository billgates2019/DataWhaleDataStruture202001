#### 1. 合并两个有序链表

#### https://leetcode-cn.com/problems/merge-two-sorted-lists/

#### 将两个有序链表合并为一个新的有序链表并返回。新链表是通过拼接给定的两个链表的所有节点组成的。

#### 示例：

#### 输入：1->2->4, 1->3->4
#### 输出：1->1->2->3->4->4

```python
class Node():
    """
    单链表中的节点应该具有两个属性：val 和 next。
    val 是当前节点的值，
    next 是指向下一个节点的指针/引用。
    """

    def __init__(self, value):
        # 存放元素数据
        self.val = value
        # next是下一个节点的标识
        self.next = None


class SingleLinkList():
    def __init__(self, node=None):
        # 头节点定义为私有变量
        self._head = node

    def travel(self):
        """
        遍历整个链表
        :return:
        """
        cur = self._head
        while cur is not None:
            print(cur.val, end=' ')
            cur = cur.next
        print()
        
        
    def insert(self, val: int) -> int:  
        """
        在一个已按从小到大排序线性表中，插入结点
        """
        node = Node(val)
        
        # 判断链表是否为空
        if self._head is None:
            self._head = node
            return
        
        # 判断是否要插入链表头
        if val<=self._head.val:
            node.next=self._head
            self._head = node
            return
        
        pre = self._head
        suc=pre.next
        while suc is not None:
            if val<=suc.val:
                break
            pre=suc
            suc=pre.next        
        
        #仅有一个结点时，或找到合适位置时
        pre.next=node
        node.next=suc



    
if __name__ == '__main__':
    
    
    L1=SingleLinkList()
    str1=input("输入L1各结点数值（以空格隔开）：")  #例如输入：3 19 4 8 4 8 2 7
    sarr=str1.split(' ')
    
    for str2 in sarr:
        val=int(str2)
        L1.insert(val)

    L1.travel()
    
    L2=SingleLinkList()
    str1=input("输入L2各结点数值（以空格隔开）：")  #例如输入：2 8 8 4 16 3 8 9 6 8 4 9 4 5
    sarr=str1.split(' ')
    
    for str2 in sarr:
        val=int(str2)
        L2.insert(val)

    L2.travel()
    
    cur = L2._head
    while cur is not None:
        L1.insert(cur.val)
        cur = cur.next
    L1.travel()
    
```

#### 2. 删除链表的倒数第N个节点

#### https://leetcode-cn.com/problems/remove-nth-node-from-end-of-list/

#### 给定一个链表，删除链表的倒数第 n 个节点，并且返回链表的头结点。

#### 示例：

#### 给定一个链表: 1->2->3->4->5, 和 n = 2.

#### 当删除了倒数第二个节点后，链表变为 1->2->3->5.

#### 说明：

#### 给定的 n 保证是有效的。

#### 进阶：

#### 你能尝试使用一趟扫描实现吗？

```python
#需要上面第1题“class Node():”段代码和“class SingleLinkList():”段代码
def length(self):
    """
    cur游标，用来移动遍历节点
    count用来计数
    :return: 返回链表的长度
    """
    cur = self._head
    count = 0
    while cur is not None:
        count += 1
        cur = cur.next
    return count
    


def delete_at_index(self, index: int) -> None or int:
    if (index<0) or (index >=self.length()):
        print("索引值超过了链表范围。")
        return -1
        
    """
    如果索引 index 有效，则删除链表中的第 index 个节点。
    :param index: 对应的索引值
    :return: -1表示为异常
     """
    pre = None
    cur = self._head
    if index is 0:
        self._head = self._head.next
        return
    for i in range(index):
        pre = cur
        cur = pre.next
    pre.next = cur.next
    
    
SingleLinkList.length=length
SingleLinkList.delete_at_index=delete_at_index


L1=SingleLinkList()
str1=input("输入L1各结点数值（以空格隔开）：")  #例如输入：1->2->3->4->5
sarr=str1.split('->')
    
for str2 in sarr:
    val=int(str2)
    L1.insert(val)

L1.travel()
    
    
len=L1.length()
revIndex=2
index=len-revIndex
    
L1.delete_at_index(index)
L1.travel()
```

#### 3. 旋转链表

#### https://leetcode-cn.com/problems/rotate-list/

#### 给定一个链表，旋转链表，将链表每个节点向右移动k个位置，其中k是非负数。

#### 示例 1:

#### 输入: 1->2->3->4->5->NULL, k = 2
#### 输出: 4->5->1->2->3->NULL

#### 解释:
#### 向右旋转 1 步: 5->1->2->3->4->NULL
#### 向右旋转 2 步: 4->5->1->2->3->NULL

#### 示例 2:

#### 输入: 0->1->2->NULL, k = 4
#### 输出: 2->0->1->NULL

#### 解释:
#### 向右旋转 1 步: 2->0->1->NULL
#### 向右旋转 2 步: 1->2->0->NULL
#### 向右旋转 3 步: 0->1->2->NULL
#### 向右旋转 4 步: 2->0->1->NULL

```python
class Node():
    def __init__(self, value):
        # 存放元素数据
        self.val = value
        # next是下一个节点的标识
        self.next = None


class SglCycLinkedlist():
    def __init__(self, node=None):
        # 头节点定义为私有变量
        self._head = None
        
    def append(self, val: int) -> None or int:
        """
        在尾部添加一个节点
        :param item:
        :return:
        """
        node = Node(val)
        # 若链表为空，直接将该节点作为链表的第一个元素
        if self._head is None:
            self._head = node
            node.next=self._head
            return
        
        cur = self._head
        while cur.next!= self._head:
            cur = cur.next
        cur.next = node
        node.next=self._head


        
    def travel(self):
        """遍历链表"""
        if self._head == None:
            print("链表为空")
            return
        
        cur = self._head
        print(cur.val,end='')
        while cur.next != self._head:
            cur = cur.next
            print(cur.val,end='')
        print()
        
        
    def rotate(self):
        """旋转1步"""
        if self._head==None:
            return
        
        cur = self._head.next
        while cur.next != self._head:
            cur = cur.next
        self._head=cur
                
    
if __name__ == '__main__':
    scll=SglCycLinkedlist()
    str1=input("输入L1各结点数值（以->隔开）：")  #例如输入：1->2->3->4->5 或 0->1->2
    sarr=str1.split('->')
    
    for str2 in sarr:
        val=int(str2)
        scll.append(val)

    scll.travel()
    
    str1=input("输入移动位置个数：")  
    k=int(str1)
    for i in range(k):
        scll.rotate()
    scll.travel()
```
