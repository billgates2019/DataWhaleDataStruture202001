<font color=brown size=4 face=雅黑>**1. 无重复字符的最长子串**</font>
#### https://leetcode-cn.com/problems/longest-substring-without-repeating-characters/

#### 给定一个字符串，请你找出其中不含有重复字符的最长子串的长度。

#### 示例1

#### 输入: "abcabcbb"
#### 输出: 3 
#### 解释: 因为无重复字符的最长子串是 "abc"，所以其长度为 3。

#### 示例2

#### 输入: "bbbbb"
#### 输出: 1
#### 解释: 因为无重复字符的最长子串是 "b"，所以其长度为 1。

#### 示例3

#### 输入: "pwwkew"
#### 输出: 3
#### 解释: 因为无重复字符的最长子串是 "wke"，所以其长度为 3。
#### 请注意，你的答案必须是 子串 的长度，"pwke" 是一个子序列，不是子串。


```python
def maxlen(str):
    if str=="":
        return 0
    
    maxlenth=1
    maxsub=str[0]
    
    
    i=0
    '''==================================i循环====================================================================='''
    while i<len(str):
        #试取子串开头字符
        tmpstr=str[i]
        
        repeatedFlag=False
        #repeatedFlag=False表示：未在子串中检测到相同字符
        
        j=i+1
        '''*******************************j循环**********************************************'''
        while not(repeatedFlag)and(j<len(str)):
        #往子串尾部添加不重复字符str[j]。
            c=str[j]
            
            
            
            
            k=i
            '''----------------------------k循环--------------------------------'''
            while not(repeatedFlag)and(k<=j-1):
                if c==str[k]:
                    repeatedFlag=True
                k+=1
            #检测子串中是否存在字符str[j]。
            #如果存在，子串不添加str[j]。同时记住前部相同字符的位置k-1。
            #下一个考察的子串首字符将从str[k]开始。
            '''----------------------------k循环---------------------------------'''
            
            
            
            if not(repeatedFlag):
            #经完整检测，字符c与子串前部各字符都不重复，子串可以加入字符c
                tmpstr=tmpstr+c
                if len(tmpstr)>maxlenth:
                    maxsub=tmpstr
                    maxlenth=len(maxsub)
            j+=1
        '''********************************j循环**********************************************'''
                    

        i+=1
        
        if repeatedFlag:
        #repeatedFlag=True表示：以str[i]开头的不重复子串，最长只能到str[j-1]，不能到str[j]。
        #原因是子串前部位置字符str[k-1]=str[j]。
        #需要考察的下一个子串，将以str[k]作为首字符。
        #至于以str[i+1]、str[i+2]...str[k-2]或str[k-1]开头的子串，其长度不会超过以str[i]开头的子串，可以跳过。
            i=k
        
        if maxlenth>=len(str)-i:
        #假如，当前maxlenth大于等于后续待考察子串可能的最大长度，则不必继续考察。
            break
    '''====================================i循环====================================================================='''
    
    print("最长子串：",maxsub,end="")
    print("    子串长度：",maxlenth)
    return maxlenth




if __name__ == "__main__":
    str="abcabcbb"
    print(str)
    maxlen(str)
    

    str="bbbbb"
    print(str)
    maxlen(str)
    

    str="pwwkew"
    print(str)
    maxlen(str)
    
    str=input(">>")
    str=str.replace(" ", "") #去除字符串中所有空格
    maxlen(str)
    
    
```

<font color=brown size=4 face=雅黑>**2. 串联所有单词的子串**</font>
#### https://leetcode-cn.com/problems/substring-with-concatenation-of-all-words/

#### 给定一个字符串s和一些长度相同的单词 words。找出 s 中恰好可以由 words 中所有单词串联形成的子串的起始位置。

#### 注意子串要与 words 中的单词完全匹配，中间不能有其他字符，但不需要考虑 words 中单词串联的顺序。

#### 示例1

#### 输入：
####   s = "barfoothefoobarman",
####   words = ["foo","bar"]
#### 输出：[0,9]
#### 解释：
#### 从索引 0 和 9 开始的子串分别是 "barfoo" 和 "foobar" 。输出的顺序不重要, [9,0] 也是有效答案。

#### 示例2

#### 输入：
####   s = "wordgoodgoodgoodbestword",
####   words = ["word","good","best","word"]
#### 输出：[]


```python
#以s = "wordgoodgoodgoodbestword"和words = ["word","good","best","word"]为例进行说明
#采用固定长度=4*4=16，以不同起始位置，把s按各种方式进行顺序切片：
# 1.把每一种切片得到的16字母，按顺序分成4个单词，每个单词4个字母。两个单词分别记为sseg[0]、sseg[1]、sseg[2]和sseg[3]。
# 2.匹配words[]和sseg[]，结果记录到“单词匹配表”：matchtable[]。
#   matchtable[k]记录words[k]对应的sseg单词序号。
#   假如：words[2]="best"，sseg[3]="best"，则记录matchtable[2]=True，表示words[2]已包含在sseg[0]、sseg[1]、sseg[2]和sseg[3]之中。
# 3.某种切片匹配结束，如果单词匹配表填满了True，则表示这个切片子串已串联所有words单词




def locate(s,words):
    slen=len(s)
    wordsnum=len(words)
    wordlen=len(words[0])
    
    substrNum=0
    
    i=0
    '''==================================i循环====================================================================='''
    while i<=slen-wordlen*wordsnum:
        sseg=[]
        matchtable=[None]*wordsnum
        
        '''*******************************j循环**********************************************'''
        for j in range(wordsnum):
            sseg.append(s[i+wordlen*j:i+wordlen*(j+1)])
            
            k=0
            segMatched=False
            '''----------------------------k循环--------------------------------'''
            while (segMatched==False) and (k<wordsnum):
                if sseg[j]==words[k]:
                    if matchtable[k]!=True:
                        matchtable[k]=True
                        segMatched=True
                k+=1
            '''----------------------------k循环---------------------------------'''
            if segMatched==False:
                break
        '''*******************************j循环**********************************************'''
        flag=True
        for m in range(wordsnum):
            flag=flag and matchtable[m]
        if flag:
            print(i,s[i:i+wordlen*wordsnum])
            substrNum=substrNum+1
            
            
        i+=1
    '''====================================i循环====================================================================='''
    if substrNum==0:
        print("[]")
    

        
        



if __name__ == "__main__":
    
    s = "barfoothefoobarman"
    words = ["foo","bar"]
    print("\n",s,words)
    locate(s,words)

    s = "wordgoodgoodgoodbestword"
    words = ["word","good","best","word"]
    print("\n",s,words)
    locate(s,words)
```

<font color=brown size=4 face=雅黑>**3. 替换子串得到平衡字符串**</font>
#### https://leetcode-cn.com/problems/replace-the-substring-for-balanced-string/

#### 有一个只含有'Q', 'W', 'E','R'四种字符，且长度为 n的字符串。假如在该字符串中，这四个字符都恰好出现n/4次，那么它就是一个「平衡字符串」。

#### 给你一个这样的字符串 s，请通过「替换一个子串」的方式，使原字符串 s变成一个「平衡字符串」。你可以用和「待替换子串」长度相同的任何其他字符串来完成替换。

#### 请返回待替换子串的最小可能长度。

#### 如果原字符串自身就是一个平衡字符串，则返回 0。

#### 示例1：

#### #### 输入：s = "QWER"
#### 输出：0
#### 解释：s 已经是平衡的了。

#### 示例2：

#### 输入：s = "QQWE"
#### 输出：1
#### 解释：我们需要把一个 'Q' 替换成 'R'，这样得到的 "RQWE" (或 "QRWE") 是平衡的。

#### 示例3：

#### 输入：s = "QQQW"
#### 输出：2
#### 解释：我们可以把前面的 "QQ" 替换成 "ER"。 

#### 示例4：

#### 输入：s = "QQQQ"
#### 输出：3
#### 解释：我们可以替换后 3 个 'Q'，使 s = "QWER"。

```python
#以16个字母为例进行说明。
#假设s = "O  D  H  K  P  C  L  D  L  H  K  A  P  K  P  G"，字母集charset = ['A', 'B', 'C', ... 'N', 'O', 'P']


#从左到右，标记s中重复出现的字母：s = "O   D   H   K   P   C   L   D   L   H   K   A   P   K   P   G"
#                                                                D2  L2  H2  K2      P2  K3  P3
#可以看到，需要替换的子串从D2到P3，长度为8


#从右到左，标记s中重复出现的字母：s = "O   D   H   K   P   C   L   D   L   H   K   A   P   K   P   G"
#                                        D2  H2  K3  P3      L2              K2      P2
#可以看到，需要替换的子串从D2到P2，长度为12


#显然，待替换子串的最小可能长度为8






import random


def substr(str,charset):
    print("\n",str,len(str))
    #print(charset,len(charset))
    
    slen=len(str)
    
    LtoR=[None]*slen
    LtoRbegin=slen
    LtoRend=-1
    
    i=0
    #从左到右，标记s中重复出现的字母
    '''==================================i循环====================================================================='''
    while i<=(slen-2):
        c=str[i]
        
        j=i+1
        #检查标记后续每个字符是否与str[i]相同
        '''*******************************j循环**********************************************'''
        while j<=(slen-1):
            if c==str[j]:
                if LtoR[j]==None:
                    LtoR[j]=2
                    if LtoRbegin>j:
                        LtoRbegin=j
                    if LtoRend<j:
                        LtoRend=j
                        
            j+=1        
        '''*******************************j循环end*******************************************'''
        
        i+=1
        while LtoR[i]!=None and i<=(slen-3):
            i+=1
    '''==================================i循环end=================================================================='''
    
    #print(LtoR)
    #print(LtoRbegin,LtoRend)
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    RtoL=[None]*slen
    RtoLbegin=slen
    RtoLend=-1
    
    i=slen-1
    #从右到左，标记s中重复出现的字母
    '''==================================i循环====================================================================='''
    while i>=1:
        c=str[i]
        
        j=i-1
        #检查标记前方每个字符是否与str[i]相同
        '''*******************************j循环**********************************************'''
        while j>=0:
            if c==str[j]:
                if RtoL[j]==None:
                    RtoL[j]=2
                    if RtoLbegin>j:
                        RtoLbegin=j
                    if RtoLend<j:
                        RtoLend=j
                        
            j-=1        
        '''*******************************j循环end*******************************************'''
        
        i-=1
        while RtoL[i]!=None and i>=2:
            i-=1
    '''==================================i循环end=================================================================='''
    
    #print(RtoL)
    #print(RtoLbegin,RtoLend)
    
    
    LtoRlen=LtoRend-LtoRbegin+1
    RtoLlen=RtoLend-RtoLbegin+1
    if LtoRlen<0:
        print(0,"平衡字符串")
        
    else:
        print("待替换子串的最小可能长度：",min(LtoRlen,RtoLlen))
    
                        

        
    
    
    
    
    

if __name__ == "__main__":
    
    charset=["Q","W","E","R"]
    
    s = "QWER"
    substr(s,charset)
    
    s = "QQWE"
    substr(s,charset)
    
    s = "QQQW"
    substr(s,charset)
    
    s = "QQQQ"
    substr(s,charset)
    
    
    
    
    charset=['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L', 'M', 'N', 'O', 'P']
    s="ODHKPCLDLHKAPKPG"
    substr(s,charset)
    
    charset=[]
    s=""
    sum=2+random.randint(0,24)
    for i in range(sum):
        charset.append(chr(65+i))
        s=s+chr(random.randint(65,65+sum-1))
    substr(s,charset)
```
